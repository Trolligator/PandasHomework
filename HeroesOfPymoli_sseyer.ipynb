```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# df_players = purchase_data['SN'].unique()
# df_players = len(df_players)
# df_players.head()

unique_players = purchase_data['SN'].unique()
num_of_players = len(unique_players)
df_total_players = pd.DataFrame({"Total Players":[num_of_players]})
df_total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].unique()
num_unique_items = len(unique_items)

average_price = purchase_data["Price"].mean()
total_sales = purchase_data["Price"].sum()
average_age = purchase_data["Age"].mean()

summary_dict = {"Average Price": [average_price], "Total Sales": [total_sales], "Average Age": [average_age],
               "Unique Items": [num_unique_items]}

df_summary = pd.DataFrame(data=summary_dict)
df_summary["Average Price"] = df_summary["Average Price"].map("{:,.2f}".format)
df_summary["Total Sales"] = df_summary["Total Sales"].map("{:,.2f}".format)
df_summary["Average Age"] = df_summary["Average Age"].round()
df_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# count_genders = unique_players
# count_genders = purchase_data["Gender"].value_counts()
count_genders = purchase_data.drop_duplicates("SN")["Gender"].value_counts()
gender_percentage = count_genders / num_of_players

df_genders = pd.DataFrame({"Count": count_genders, "Percentage of Players": gender_percentage})
df_genders["Percentage of Players"] = df_genders["Percentage of Players"].map("{:.2%}".format)
df_genders
```

```python
# Grouping purchase data by gender and calculating new columns
grouped_purchase_data = purchase_data.groupby("Gender")
total_purchase_grouped = grouped_purchase_data["Price"].sum()
mean_purchase_grouped = grouped_purchase_data["Price"].mean()
average_purchase_grouped = total_purchase_grouped / count_genders
count_gender_purchase_data = grouped_purchase_data["Gender"].count()

# Creating new dataframe for analysis by gender
df_gender_analysis = pd.DataFrame({"Total Purchases":total_purchase_grouped,
                                   "Average Purchase":mean_purchase_grouped,
                                   "Purchases by Gender":count_gender_purchase_data,
                                   "Average Purchase per Person":count_gender_purchase_data})

# Formatting columns
df_gender_analysis["Average Purchase"]=df_gender_analysis["Average Purchase"].map("${:,.2f}".format)
df_gender_analysis["Purchases by Gender"]=df_gender_analysis["Purchases by Gender"].map("${:,.2f}".format)
df_gender_analysis["Average Purchase per Person"]=df_gender_analysis["Average Purchase per Person"].map("${:,.2f}".format)

df_gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_labels)
df_unique = purchase_data.drop_duplicates("SN")

count_age = df_unique.groupby("Age Group")["SN"].count()
percentage_age = count_age / num_of_players

df_age = pd.DataFrame({"Amount of Players":count_age,
                      "Percentage of Players":percentage_age})
df_age["Percentage of Players"]=df_age["Percentage of Players"].map("{:.2%}".format)

df_age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_age = purchase_data.groupby("Age Group")
purchase_age_count = purchase_age["Purchase ID"].count()
purchase_age_mean = purchase_age["Price"].mean()
purchase_age_sum = purchase_age["Price"].sum()
purchase_age_average = purchase_age_sum / purchase_age_count

df_age_analysis = pd.DataFrame({"Number of Purchases":purchase_age_count,
                               "Average Purchase Price": purchase_age_mean,
                               "Sum of Purchases":purchase_age_sum,
                               "Average per Person":purchase_age_average
                               })

df_age_analysis["Average Purchase Price"]=df_age_analysis["Average Purchase Price"].map("${:,.2f}".format)
df_age_analysis["Sum of Purchases"]=df_age_analysis["Sum of Purchases"].map("${:,.2f}".format)
df_age_analysis["Average per Person"]=df_age_analysis["Average per Person"].map("${:,.2f}".format)

df_age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_SN = purchase_data.groupby("SN")

spenders_count = spenders_SN["Purchase ID"].count()
spenders_total = spenders_SN["Price"].sum()
spenders_mean = spenders_SN["Price"].mean()

df_top_spenders = pd.DataFrame({"Purchase Count":spenders_count,
                               "Total Purchase Value":spenders_total,
                               "Average Purchase Price":spenders_mean
                               })

df_top_spenders = df_top_spenders.sort_values(by="Total Purchase Value", ascending=False)
df_top_spenders["Total Purchase Value"] = df_top_spenders["Total Purchase Value"].map("${:,.2f}".format)
df_top_spenders["Average Purchase Price"] = df_top_spenders["Average Purchase Price"].map("${:,.2f}".format)

df_top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_values = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
item_values = item_values.groupby(["Item ID", "Item Name"])

item_count = item_values["Item ID"].count()
item_price = item_values["Price"].mean()
item_total_price = item_values["Price"].sum()

df_popular = pd.DataFrame({"Purchase Count":item_count,
                          "Item Price":item_price,
                          "Total Purchase Value":item_total_price
                          })
df_popular = df_popular.sort_values(by="Purchase Count", ascending=False)

df_popular["Item Price"] = df_popular["Item Price"].map("${:,.2f}".format)
df_popular["Total Purchase Value"] = df_popular["Total Purchase Value"].map("${:,.2f}".format)

df_popular.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = df_popular.sort_values(by="Total Purchase Value", ascending=False)



most_profitable.head()
```
